```python
#time to clean starting with anime data
def clean_anime(anime_data) -> pd.DataFrame:
    anime_data.columns = anime_data.columns.str.title()
    anime_data.columns = anime_data.columns.str.strip().str.replace('_', ' ')

    # Drop unnecessary columns
    cols_to_drop = ['episodes']
    anime_data = anime_data.drop(columns=cols_to_drop)

    # Drop rows with missing values
    anime_data = anime_data.dropna()

    return anime_data
```

```python
#time to clean starting with anime rating
def clean_anime_rating(anime_rating) -> pd.DataFrame:
    anime_rating.columns = anime_rating.columns.str.title()
    anime_rating.columns= anime_rating.columns.str.strip().str.replace('_', ' ')

    # Drop unnecessary columns - not needed in this case
    

    # Drop rows with missing values
    anime_rating = anime_rating.dropna()

    return anime_rating
```

```python
# Call function clean

anime_data_clean = clean_anime(anime_data)
anime_data_clean.head(5)
```

```python
anime_rating_clean = clean_anime_rating(anime_rating)
anime_rating_clean.head(5)
```
