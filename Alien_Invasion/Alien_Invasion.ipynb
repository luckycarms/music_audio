# Alien Invasion Game

In *Alien Invasion*, the player controls a rocket ship that appears at the bottom center of the screen. The player can move the ship right and left using the arrow keys and shoot bullets using the spacebar. When the game beins, a fleet of aliens fills the sky and moves across and down the screen. The player shoots and destroys aliens. If the player destroys all the aliens, a new fleet appears that moves faster than the previous fleet. If any alien hits the player's ship or reaches the bottom of the screen, they player loses a ship. If the player loses three ships, the game ends.

Matthews, Eric. Python Crash Course. San Fransisco, no starch press, 2015.
